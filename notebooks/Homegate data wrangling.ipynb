```python
def remove_unwanted_char(org_str):
    mod_string = ""
    for i in range(3, len(org_str)):
        mod_string = mod_string + org_str[i]
    return mod_string[:-2]

df['price'] = df['price'].apply(remove_unwanted_char)
df
```

```python
df['area_m2'] = df['area_m2'].str.split(' ').str[0]
```

```python
df[['floors_num']] = df[['floors_num']].fillna(1)
```

```python
def extract_zip(address):
    str1 = [s for s in address.split() if not s.isdigit()]
    return " ".join(str1)

#need to extract city to attach to certain canton
def city(address):
    if ',' in address:
        str1 = address.split(',')
        str2 = str1[1]
    else:
        str2 = address
    return str2
```

```python
df['zip_city'] = df['address'].apply(city)
```

```python
df['city'] = df['zip_city'].apply(extract_zip)
```

```python
def extract_zip2(address):
    str1 = [s for s in address.split() if s.isdigit()]
    str2 = ""   
    for ele in str1: 
        str2 += ele
    return str2
```
